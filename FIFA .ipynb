#### Handling outlier in 'age'

```python
upper_limit = data.age.mean() + 3*data.age.std()
print("Upper limit:",upper_limit)
lower_limit = data.age.mean() - 3*data.age.std()
print("Lower limit:",lower_limit)
```

```python
# Getting the percentage of outlier
len(data.loc[data['age'] > upper_limit])/18278*100
# In age feature less than 5% outlier we are impute the outlier
```

```python
# impute outlier with mean because feature follow a normal distribution and continous data
data.loc[data['age'] > upper_limit,'age']=data['age'].mean()
```

```python
#overall rating vs age
average_ratings_by_age = data.groupby('age')['overall'].mean().reset_index()

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='overall', data=average_ratings_by_age, marker='o')
plt.title('Average Overall Rating vs. Age')
plt.xlabel('Age')
plt.ylabel('Average Overall Rating')
plt.grid(True)
```

### solution to task:Which type of offensive players tends to get paid the most: the striker, the right-winger, or the left-winger?

```python
# Filter the data for the specific positions: striker, right-winger, and left-winger
positions = ['ST', 'RW', 'LW']
filtered_data = data[data['team_position'].isin(positions)]

# Calculate the average salary for each position
average_salary_by_position = filtered_data.groupby('team_position')['value_eur'].mean().reset_index()

# Create a bar plot to visualize the average salaries
plt.figure(figsize=(10, 6))
sns.barplot(x='team_position', y='value_eur', data=average_salary_by_position, palette='Set2')
plt.title('Average Salary by Offensive Position')
plt.xlabel('Position')
plt.ylabel('Average Salary')
plt.grid(True)

plt.show()
```

#### Average Salaries by Position
Left-Winger:
The average salary for players in the Left-Winger position within the defensive category is approximately 68,000 Euros. This position typically requires players to contribute to both the attack and defense, using their speed and skills to support the team's overall strategy.

Right-Winger:
Right-Wingers in the defensive category earn an average salary of around 71,000 Euros. These players play a crucial role in the team's defensive strategy, utilizing their abilities to provide coverage and support to the backline.

Striker:
Strikers in the defensive category have an average salary of about 49,000 Euros. Despite being primarily associated with offensive play, some strikers are known for their defensive contributions, helping to disrupt the opponent's attacks.

Interpretation
The data suggests variations in average salaries for different positions within the defensive category. Right-Wingers in the defensive position tend to earn the highest average salary, possibly due to their dual roles in defense and attack. Their responsibilities often require versatility and tactical understanding, which may be reflected in their salaries.

Left-Wingers in the defensive position also earn a competitive average salary, indicating the value of their contributions on both sides of the field.

Strikers in the defensive category receive a comparatively lower average salary. While their primary role is offensive, some strikers are known for their defensive work. The variation in salaries may be attributed to differences in playing styles and individual player attributes.

## K mean clustering using elbow technique

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,10):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,10),wcss,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## silhouette_score

```python
#Import KMeans Algorithm from Scikit-learn library
from sklearn.cluster import KMeans


for i in range(2,6):

  model = KMeans(n_clusters=i, random_state=42)

#Training the model
  model.fit(X)

  label=model.labels_
  label

  from sklearn.metrics import silhouette_score
  score=silhouette_score(X,label)
  print('The silhouettee_score for {} clusters in KMeans Algorithm is {}.'.format(i,round(score,2)))
```

# Model Comparison Report

## Overview

In this report, we will compare the performance of the K-Means clustering algorithm with varying numbers of clusters (K) based on the silhouette score. The silhouette score is used to evaluate the quality of clustering, with higher values indicating better-defined clusters.

## Experimental Setup

The experiments were conducted using the K-Means clustering algorithm implemented in scikit-learn. The silhouette score was computed for K values ranging from 2 to 5.

## Results

The silhouette scores for different values of K are as follows:

- For 2 clusters in K-Means Algorithm: 0.89
- For 3 clusters in K-Means Algorithm: 0.82
- For 4 clusters in K-Means Algorithm: 0.79
- For 5 clusters in K-Means Algorithm: 0.76

## Analysis

1. Silhouette Score Trend: The silhouette score decreases as the number of clusters (K) increases. This is a common trend in K-Means clustering. The silhouette score measures the cohesion and separation of clusters, and a higher value indicates better-defined clusters. Therefore, a silhouette score of 0.89 for 2 clusters suggests that 2 clusters provide the most well-defined separation among data points.

2. Optimal Number of Clusters: The optimal number of clusters depends on the specific problem and the trade-off between data segmentation and interpretability. Based on the silhouette scores, it appears that 2 clusters result in the highest silhouette score, indicating that this may be the best choice for this dataset. However, it's important to consider the domain context and how the clustering results will be used.


## Conclusion

The choice of the number of clusters (K) in the K-Means clustering algorithm is a critical decision that impacts the quality of clustering. In this experiment, a silhouette score analysis indicates that 2 clusters provide the highest silhouette score, suggesting better-defined clusters. However, the choice of K should not be based solely on silhouette scores but should consider the domain-specific context and the intended use of the clustering results.
